```python
%%sql
UPDATE crimes_restricted
SET Categorization =
    CASE 
        WHEN Category IN ('ARSON', 'BURGLARY','EMBEZZELMENT', 'LARCENY/THEFT', 'STOLEN PROPERTY',
                        'VANDALISM', 'VEHICLE THEFT') THEN 'Property crime'
        WHEN Category IN ('ASSAULT','KIDNAPPING', 'SEX OFFENSES FORCIBLE', 'SEX OFFENSES NON FORCIBLE') 
        THEN 'Violent crime'
        WHEN Category = 'ROBBERY' THEN 'Robbery'
        ELSE 'Other'
    END
```

```python
start = time.time()

evictionsDict = {}
for i in range(0,len(census)):
    print("Working on evicitons_" + str(int(census[i]*100)))
    c.execute('''
        SELECT Date, Evict_Count, census_tr, non_payment, illegal_use, eviction, Category
        FROM days 
        LEFT OUTER JOIN (SELECT date as Dates, count(strftime('%Y %m %d', date)) AS Evict_Count, census_tr, 
        non_payment, illegal_use, eviction, Category
                 FROM evictions_''' + str(int(census[i]*100)) + '''
                 GROUP by strftime('%Y %m %d', date)) 
                 ON strftime('%Y %m %d', Dates) = strftime('%Y %m %d', Date) ''')
    print("Fetching data...")
    evictions_pd = pd.DataFrame(data = c.fetchall(), columns = ['Date', 'Evict_Count', 'census_tr', 'non_payment', 
                                                   'illegal_use', 'eviction', 'Category'])
    print("Fetching!")
    evictionsDict['evictions_{}'.format(int(census[i]*100))] = evictions_pd
    
end = time.time()
print("Time to run:", end-start, "seconds")
```

```python
start = time.time()

for evictions in evictionsDict:
    evictionsDict[evictions]=evictionsDict[evictions].fillna(0)
    for d in [1, 2, 3, 7, 14]:
        print("Calculations for " + evictions + ": " + str(d) + " days")
        last_days = [0]*d
        for i in range(d+1, evictionsDict[evictions].shape[0]+1):
            last_days.append(sum(evictionsDict[evictions]['Evict_Count'][int(i-d-1):int(i-1)]))
        evictionsDict[evictions]['last_'+str(d)+'days'] = last_days
    print(evictionsDict[evictions][36:41])
end = time.time()
print("Time to run:", end-start, "seconds")
```
