```python
import geopandas as gpd
import os

# Read in csv file containing SF neighborhoods and corresponding MULTIPOLYGON objects
data_path = '/Users/Sam/Desktop/project1030/SF'
sf_csv = pd.read_csv(os.path.join(data_path, 'San_Francisco_Analysis_Neighborhoods.csv'))
```

```python
from geopandas import GeoDataFrame
import shapely.wkt

# Create geodataframe called sf_map containing SF neighborhood boundaries
geometry = sf_csv['the_geom'].map(shapely.wkt.loads)
sf_csv = sf_csv.drop('the_geom', axis=1)

print(sf_csv.head())
crs = {'init': 'epsg:4326'}
sf_map = GeoDataFrame(sf_csv, crs=crs, geometry=geometry)
```

```python
# Create geodataframe for 2010 census tracts
census_tr = pd.read_csv(os.path.join(data_path, 'Census_2010_Tracts.csv'))

geometry = census_tr['the_geom'].map(shapely.wkt.loads)
census_tr = census_tr.drop('the_geom', axis=1)
print(census_tr.head())

# Restrict to mainland SF
census_tr = census_tr[census_tr['INTPTLON10'] > -122.6]

sf_census_tracts = GeoDataFrame(census_tr, crs=crs, geometry=geometry)
```

```python
sf_map.plot(figsize=(10, 20))
plt.show()
```

```python
sf_census_tracts.plot(figsize=(10, 20))
plt.show()
```

```python
import re

# Write function to convert "(x, y)" to x and y columns 
def convert_coords(coords):
    m = re.split(',', coords)
    x_trim = m[0].replace("(", '')
    x_num = float(x_trim)

    y_trim = m[1].replace(")", '')
    y_num = float(y_trim)
    return (x_num, y_num)

x_coords = [i[1] for i in list(map(convert_coords, evict['Location']))]
y_coords = [i[0] for i in list(map(convert_coords, evict['Location']))]

evict['X'] = x_coords
evict['Y'] = y_coords
```

```python
# Create GeoDataFrame for 2003 evictions
evict_2003 = evict[evict['year'] == 2003]
evict_locs = [Point(xy) for xy in zip(evict_2003['X'], evict_2003['Y'])]
evict_locs_df = GeoDataFrame(evict_locs, crs=crs, geometry=evict_locs)
```

```python
# Explore correlation between evictions and crime in Mission district
crime_miss = crime[crime['PdDistrict'] == "MISSION"]
evict_miss = evict[evict['Neighborhood'] == "Mission"]

print(crime_miss.shape)
print(evict_miss.shape)
```

```python
crime_miss.sort_values(by='Dates', inplace=True)
evict_miss.sort_values(by='Date', inplace=True)

#print(crime_miss.head())
#print(evict_miss.head())

print(crime_miss['Dates'].tail())   # Dates from 01-2003 to 05-2015
print(evict_miss['Date'].head())    # Dates from 01-1997 to 09-2017
```

```python
# Restrict eviction data to dates for which we have crime data
mask = (evict_miss['Date'] > '2003-1-6') & (evict_miss['Date'] <= '2015-5-13')
evict_miss_trim = evict_miss.loc[mask]
```

```python
# Now, let's create a variable in each dataset that contains the number of crimes/evictions per month
months_only = crime_miss['date'].map(lambda t: t.month)
crime_miss['month'] = months_only

months_only_e = evict_miss_trim['Date'].map(lambda t: t.month)
evict_miss_trim['month'] = months_only_e
```

```python
crime_miss['idx'] = range(1, len(crime_miss) + 1)
print(crime_miss.head())
```

```python
evict_counts = evict_miss_trim.groupby(["month", "year"])['Eviction ID'].count().reset_index(name="evictions_per_month")
crime_counts = crime_miss.groupby(["month", "year"])['idx'].count().reset_index(name="crimes_per_month")
```

```python
my_counts_merged = pd.merge(crime_counts, evict_counts, how='inner', on=['month', 'year'])
```

```python
print(my_counts_merged)
my_counts_merged.to_csv("~/Desktop/project1030/SF/exploratory/lin_reg_data.csv")
```

```python
import statsmodels.formula.api as sm
```

```python
result = sm.ols(formula="crimes_per_month ~ evictions_per_month", data=my_counts_merged).fit()
print(result.summary())
```
